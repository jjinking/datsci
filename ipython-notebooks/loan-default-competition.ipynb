```python
# Imput and standardize
_df = munge.impute_standardize(train_scaled_infocol_defaultcol_matched,
                               cols=train_features_cols)
train_scaled_infocol_defaultcol_matched_imputed_norm = _df
```

```python
# Cluster the columns based on correlations
_df = train_scaled_infocol_defaultcol_matched_imputed_norm
col_corr_clusters = eda.get_feature_clusters(_df,
                                             cols=train_features_cols,
                                             thresh=0.95,
                                             method='pearson')
```

```python
_df = train_scaled_infocol_defaultcol_matched_imputed_norm
cols_ranked, importances = eda.rank_order_features(_df[train_features_cols],
                                                   _df[train_labels_cols],
                                                   plot=True)
```

```python
gen = eda.generate_important_features(col_corr_clusters, cols_ranked)
```

### Random Forest 10-Fold CV

```python
_df = train_scaled_infocol_defaultcol_matched_imputed_norm
feature_groups = [['f527','f528'],
                  gens[:5], gens[:10], gens[:20], gens[:50],
                  gens[:100], gens[:150], gens[:200], gens]
titles = ['f527 + f528','top5','top10','top20','top50','top100','top150','top200','all']
```

```python
clf = RandomForestClassifier(n_estimators=200)
eda.cross_validate_feature_groups(clf, _df, feature_groups, _df[train_labels_cols],
                                  titles=titles, cv=10, plot=True, plots_per_row=3)
```

### SVC Forest 10-Fold CV

```python
_df = train_scaled_infocol_defaultcol_matched_imputed_norm
feature_groups = [['f527','f528'],
                  gens[:5], gens[:10], gens[:20], gens[:50],
                  gens[:100], gens[:150], gens[:200], gens]
titles = ['f527 + f528','top5','top10','top20','top50','top100','top150','top200','all']
```

```python
clf = svm.SVC()
eda.cross_validate_feature_groups(clf, _df, feature_groups, _df[train_labels_cols],
                                  titles=titles, cv=10, plot=True, plots_per_row=3)
```

### Logistic Regression (GLM w Binomial)

```python
clf = linear_model.LogisticRegression()
eda.cross_validate_feature_groups(clf, _df, feature_groups, _df[train_labels_cols],
                                  titles=titles, cv=10, plot=True, plots_per_row=3)
```

### Logistic Regression on matched data, w/o binary

```python
# Get matched counts
train_scaled_infocol_matched = train_scaled_infocol.ix[train_scaled_infocol_defaultcol_matched_imputed_norm.index]

# Imput and standardize
train_scaled_infocol_matched_imputed_norm = munge.impute_standardize(train_scaled_infocol_matched,
                                                                     cols=train_features_cols)
```

```python
_df = train_scaled_infocol_matched_imputed_norm
clf = linear_model.LogisticRegression()
eda.cross_validate_feature_groups(clf, _df, feature_groups, _df[train_labels_cols],
                                  titles=titles, cv=10, plot=True, plots_per_row=3)
```
